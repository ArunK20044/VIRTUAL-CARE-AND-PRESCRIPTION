```python
X_train = data_train.drop('prognosis', axis=1)
y_train = data_train['prognosis']
X_test = data_test.drop('prognosis', axis=1)
y_test = data_test['prognosis']

# Binarize the labels for one-vs-rest classification
lb = LabelBinarizer()
y_train_binarized = lb.fit_transform(y_train)
y_test_binarized = lb.transform(y_test)

# Step 2: Train One-vs-Rest Gradient Boosting Models
models = {}
metrics = {}

for i, disease in enumerate(lb.classes_):
    y_train_disease = y_train_binarized[:, i]
    y_test_disease = y_test_binarized[:, i]
    
    model = GradientBoostingClassifier()
    model.fit(X_train, y_train_disease)
    
    y_pred = model.predict(X_test)
    y_pred_proba = model.predict_proba(X_test)[:, 1]
    
    accuracy = accuracy_score(y_test_disease, y_pred)
    roc_auc = roc_auc_score(y_test_disease, y_pred_proba)
    cm = confusion_matrix(y_test_disease, y_pred)
    cr = classification_report(y_test_disease, y_pred)
    
    models[disease] = model
    metrics[disease] = {
        'accuracy': accuracy,
        'roc_auc': roc_auc,
        'confusion_matrix': cm,
        'classification_report': cr
    }
    
    print(f"Model for disease '{disease}'")
    print(f"Accuracy: {accuracy:.2f}")
    print(f"ROC AUC: {roc_auc:.2f}")
    print("Confusion Matrix:")
    print(cm)
    print("Classification Report:")
    print(cr)
    print("-" * 50)

```

```python
import pickle
import matplotlib.pyplot as plt
import seaborn as sns

# Function to save models as pickle files
def save_models(models, filename_prefix='model_'):
    for disease, model in models.items():
        filename = f"{filename_prefix}{disease.replace(' ', '_').lower()}.pkl"
        with open(filename, 'wb') as file:
            pickle.dump(model, file)
        print(f"Model for disease '{disease}' saved as '{filename}'")

# Save models as pickle files
save_models(models)

# Function to plot confusion matrix
def plot_confusion_matrix(y_true, y_pred, labels, title='Confusion Matrix', cmap='Blues'):
    cm = confusion_matrix(y_true, y_pred, labels=labels)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap=cmap, xticklabels=labels, yticklabels=labels)
    plt.xlabel('Predicted labels')
    plt.ylabel('True labels')
    plt.title(title)
    plt.show()

# Example Usage for plotting confusion matrix
for disease, model in models.items():
    y_pred = model.predict(X_test)
    plot_confusion_matrix(y_test_binarized[:, lb.transform([disease])[0][0]], y_pred, [0, 1], title=f"Confusion Matrix for '{disease}'")
```

```python
def set_symptoms(symptoms, num_given_symptoms):
    symptoms_all = X_train.columns.tolist()
    given_symptoms = set(symptoms[:num_given_symptoms])
    rest_symptoms = {symptom: 1 if symptom in given_symptoms else 0 for symptom in symptoms_all}
    return rest_symptoms
```

```python
def predict_disease(symptoms):
    input_data = pd.DataFrame([set_symptoms(symptoms, len(symptoms))])
    probabilities = {}
    
    for disease, model in models.items():
        prob = model.predict_proba(input_data)[0][1]  # Probability of having the disease
        probabilities[disease] = prob
    
    predicted_disease = max(probabilities, key=probabilities.get)
    
    return predicted_disease, probabilities

def get_max_probability_disease(symptoms):
    predicted_disease, probabilities = predict_disease(symptoms)
    max_probability_disease = max(probabilities, key=probabilities.get)
    max_probability = probabilities[max_probability_disease]
    
    return max_probability_disease, max_probability

# Example Usage
example_symptoms = ['itching', 'skin_rash', 'nodal_skin_eruptions', 'weight_gain']
predicted_disease, disease_probabilities = predict_disease(example_symptoms)
max_probability_disease, max_probability = get_max_probability_disease(example_symptoms)

print(f"Predicted Disease: {predicted_disease}")
print(f"Disease Probabilities: {disease_probabilities}")
print(f"Disease with Max Probability: {max_probability_disease} (Probability: {max_probability})")
```
